## Bayesian Analysis:
### What factors influence shifts in the end of season position in the English Premier League

```python
# Import necessary packages
import pandas as pd
import numpy as np
import pymc3 as pm
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from theano import shared, tensor as tt
from sklearn.preprocessing import scale
from collections import OrderedDict
import warnings
warnings.filterwarnings('ignore')

#Pandas Options
pd.set_option('display.max_rows', 50)
pd.set_option('display.max_columns', 50)
```

## Data Wrangling

```python
# Read in the raw data
results = pd.read_csv('results.csv')
results.head()
```

```python
results['home_fts'] = results['home_goals'].apply(lambda x: 1 if x==0 else 0)
results['away_fts'] = results['away_goals'].apply(lambda x: 1 if x==0 else 0)
results['home_goals_conceded'] = results['away_goals']
results['away_goals_conceded'] = results['home_goals']
results['home_wins'] = results['result'].apply(lambda x: 1 if x == 'H' else 0)
results['home_draws'] = results['result'].apply(lambda x: 1 if x == 'D' else 0)
results['home_losses'] = results['result'].apply(lambda x: 1 if x == 'A' else 0)
results['away_wins'] = results['result'].apply(lambda x: 1 if x == 'A' else 0)
results['away_draws'] = results['result'].apply(lambda x: 1 if x == 'D' else 0)
results['away_losses'] = results['result'].apply(lambda x: 1 if x == 'H' else 0)


home_columns = ['season']+[x for x in results.columns if x.find('home')>-1]
away_columns = ['season']+[x for x in results.columns if x.find('away')>-1]

home_results = results[home_columns]
away_results = results[away_columns]

home_results.rename({'home_team':'team'}, axis = 1, inplace = True)
away_results.rename({'away_team':'team'}, axis = 1, inplace = True)

home_results_summary = home_results.groupby(['team','season']).sum()
away_results_summary = away_results.groupby(['team','season']).sum()

results_df = pd.merge(home_results_summary, away_results_summary, on = ['team','season'])
results_df['fts'] = results_df['home_fts'] + results_df['away_fts']
results_df['ha_win_ratio'] = results_df['home_wins']/results_df['away_wins']

results_df.head()
```

```python
results_df.columns
```

```python
stats = pd.read_csv('stats.csv')
stats.set_index(['team','season'], inplace = True)

stats['draws'] = 38 - stats['wins'] - stats['losses']
stats['points'] = stats['wins']*3 + stats['draws']*1
stats['goal_difference'] = stats['goals'] - stats['goals_conceded']

stats.rename({'total_scoring_att':'shots',
              'ontarget_scoring_att':'shots_on_target',
              'total_through_ball':'through_balls',
              'total_pass':'passes',
              'total_long_balls':'long_balls',
              'total_cross':'crosses', 
              'corner_taken':'corners',
              'goal_fastbreak':"goals_counter"}, axis = 1, inplace = True)

stats['bookings'] = stats['total_red_card'] + stats['total_yel_card']

stats.sort_values(['season','points','goal_difference','goals'], ascending = False, inplace = True)
stats["position"] = stats.groupby("season")["points"].rank("first", ascending=False)

stats.head()
```

```python
# Creating combined Stats data to Finally use

required_cols =  ['position',
 'goals_conceded',
 'shots',
 'shots_on_target',
 'goals_counter',
 'long_balls',
 'passes',
 'touches',
 'crosses',
 'through_balls',
 'corners',
 'att_obox_goal',
 'bookings']

stats_df = results_df.join(stats)[required_cols]


position_category = {1:1, 2:1, 3:1, 4:1,
                    5:2, 6:2, 7:2, 8:2, 9:2, 10:2,
                    11:3, 12:3, 13:3, 14:3, 15:3, 16:3, 17:3,
                    18:4, 19:4, 20:4}

position_cat_desc = {1:'Champions League Qlf', 2:'Top Half', 3:'Bottom Half', 4:'Relegated'}

stats_df['position_category'] = stats_df['position'].astype('int').map(position_category)
stats_df['position_cat_desc'] = stats_df['position_category'].astype('int').map(position_cat_desc)

stats_df.head()
```

```python
stats_df.to_csv('epl_stats_2006-2018_curated.csv')
```

```python
#Checking missing values
stats_df.isnull().sum()
```

```python
# Impute missing values
stats_df['through_balls'] = stats_df['through_balls'].fillna(stats_df.groupby('position_category')['through_balls'].transform('mean')).round()
```

```python
#Creating Position Shifts Dataset
stats_cols = [x for x in required_cols if x not in ['position','season']] + ['position_category']
stats_shifts_df = stats_df[stats_cols].reset_index().groupby(['team'])[stats_cols].diff()

stats_shifts_df.rename({'position_category':'position_cat_shift'}, axis = 1, inplace = True)
stats_shifts_df['position_cat_shift'] = 0.0 - stats_shifts_df['position_cat_shift']
stats_shifts_df['position_cat_ls'] = stats_df['position_category'].values + stats_shifts_df['position_cat_shift']

#restricting shifts to a maximum of 1 standing class or position category
stats_shifts_df['position_cat_shift'].clip(lower = -1, upper = 1, inplace = True) 

stats_shifts_df.dropna(how = 'all', inplace = True)

# Dropping relegation zone stats
stats_shifts_df.drop(stats_shifts_df.loc[stats_shifts_df['position_cat_ls'] == 4].index, inplace = True)

stats_shifts_df.set_index('position_cat_shift', inplace = True)
stats_shifts_df.head()
```

## Build Models

```python
data = pd.read_csv('epl_stats_shifts.csv')
data.head()
```

```python
# Check number of observations in each standing class
data.groupby('position_cat_shift').count()
```

```python
# Define X cols, removing the correlated predictors
cols = ['goals_conceded', 
        'shots', 
        #'shots_on_target',
        'goals_counter', 
        'long_balls', 
        'passes', 
        #'touches', 
        'crosses',
        'through_balls', 
        #'corners', 
        'att_obox_goal', 
        'bookings']
```

```python
# Prep input data for model
Y = pd.Categorical(data['position_cat_shift']).codes
X = scale(data[cols].values)
```

```python
# Verify shape
X.shape
```

## Complete Pooling

```python
try:
    # augment X for simpler regression expression
    X_aug = tt.concatenate((np.ones((X.shape[0], 1)), X), axis=1)
    
    with pm.Model() as ordered_mvlogit_pooled:
        # regression coefficients
        beta = pm.Normal('beta', mu=0, sd=2, shape=X.shape[1] + 1)

        # transformed space (univariate real)
        eta = X_aug.dot(beta)

        # hyperplanes for separating categories
        cutpoints = pm.Normal("cutpoints", mu=[-1,1], sd=1, shape=2,
                              transform=pm.distributions.transforms.ordered)

        y_ = pm.OrderedLogistic("y", cutpoints=cutpoints, eta=eta, observed=Y)
except:
    # augment X for simpler regression expression
    X_aug = tt.concatenate((np.ones((X.shape[0], 1)), X), axis=1)
    
    with pm.Model() as ordered_mvlogit_pooled:
        # regression coefficients
        beta = pm.Normal('beta', mu=0, sd=2, shape=X.shape[1] + 1)

        # transformed space (univariate real)
        eta = X_aug.dot(beta)

        # hyperplanes for separating categories
        cutpoints = pm.Normal("cutpoints", mu=[-1,1], sd=1, shape=2,
                              transform=pm.distributions.transforms.ordered)

        y_ = pm.OrderedLogistic("y", cutpoints=cutpoints, eta=eta, observed=Y)
```

## Partial Pooling: Variable Intercepts

```python
# Additional data prep for the varying intercepts
ls_pos = data['position_cat_ls'].values.astype('int')
unique_positions = data['position_cat_ls'].nunique()
X_tensor = tt._shared(X)
```

```python
# Model with sampling
with pm.Model() as ordered_mvlogit_var_int:
    
    # Intercepts for separate league positions last season
    mu_beta_0 = pm.Normal('mu_beta_0', mu=0, sd=2)
    sd_beta_0 = pm.HalfCauchy('sd_beta_0', beta=2)
    
    beta_0 = pm.Normal('beta_0', mu = mu_beta_0, sd = sd_beta_0, shape=unique_positions)
    
    # regression coefficients
    sd_beta = pm.HalfCauchy('sd_beta', beta=2)
    beta = pm.Normal('beta', mu = 0, sd = sd_beta, shape=X.shape[1])

    # transformed space (univariate real)
    eta = X_tensor.dot(beta) + beta_0[ls_pos-1]

    # hyperplanes for separating categories
    cutpoint_sigma=pm.HalfNormal('cutpoint_sigma', sigma = 1)
    cutpoints = pm.Normal("cutpoints", mu=np.array([-1,1]), sd=cutpoint_sigma, shape=2,
                          transform=pm.distributions.transforms.ordered)

    y_ = pm.OrderedLogistic("y", cutpoints=cutpoints, eta=eta, observed=Y)
```

```python
pm.model_to_graphviz(ordered_mvlogit_var_int)
```

```python
with ordered_mvlogit_var_int:
    footy_trace_var_int = pm.sample(20000, tune = 1000, cores = 4)
```

## Compare Models

```python
# Get models and traces for waic comparison
model_lst = [ordered_mvlogit_pooled, ordered_mvlogit_var_int, footy_trace_pooled_int]
traces_lst = [footy_trace_pooled, footy_trace_var_int, footy_trace_pooled_int]

def prepare_models(model_lst):
    traces = OrderedDict()
    
    for m in range(len(model_lst)):
        name = 'm{}'.format(m)
        print('\nPreparing: {}'.format(name))
        traces[name] = traces_lst[m]
    return traces

traces_lin = prepare_models(model_lst)
```

```python
model_trace_dict = dict()
for i, name in enumerate(list(traces_lin.keys())):
    model_trace_dict.update({name: traces_lin[name]})
waic_df = pm.compare(model_trace_dict, ic='WAIC')
waic_df
```
